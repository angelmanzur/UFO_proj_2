```python
import numpy as np
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
import time
import csv
```

```python
# open the base page and get the table with links of sightings per month
```

```python
base_url = "http://www.nuforc.org/webreports/"

ufo_url = base_url + "ndxevent.html"
html_page = requests.get(ufo_url)
soup = BeautifulSoup(html_page.content, 'html.parser')
```

```python
# get the tabla, 
```

```python
table = soup.find('table')
```

# get the elements of the table
The format is 
   
       link      count
       link      count


```python
ufolinks = table.find_all('td')
```

```python
new_url = []
for ufolink in ufolinks:
    try:
        nextlink = ufolink.find('a').attrs['href']
        new_url.append(nextlink) 
    except:
        #not a link, move on
        pass
new_url.pop() # remove the last link as it is a link to main page
```

```python
total_links = len(new_url)
total_links
```

```python
ufo_header =['Date_Time', 'City','State','Shape','Duration','Summary','Date_Posted' ]
```

```python
ufo_reports = []
ufo_reports.append(ufo_header)
nlinks = 1
for url in new_url:
    ufo_url = base_url + url
    print(ufo_url)
    
    try:
        ufo_page = requests.get(ufo_url)
        soup = BeautifulSoup(ufo_page.content, 'html.parser')
        events = soup.find('table').find('tbody').find_all('tr')

        for event in events:
            ufo_lst = event.find_all('td')
            ufo_row = []
            for entry in ufo_lst:

                ufo_row.append(entry.text)
            ufo_reports.append(ufo_row)

    except:
        pass

    if nlinks%60 == 0:
        try:
            with open('ufo.csv', 'w', newline='\n') as myfile: 
                wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
                for ufos in ufo_reports:
                    wr.writerow(ufos)
        except:
            #nothing to do
            pass
    
    nlinks += 1
    
    print(nlinks)
    time.sleep(2)
    #sleep 2 seconds between the link so 
#     if nlinks == total_links:
#         break; #th
```

```python
# save all the date one last time
with open('ufo.csv', 'w', newline='\n') as myfile: 
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    for ufos in ufo_reports:
        wr.writerow(ufos)
```
